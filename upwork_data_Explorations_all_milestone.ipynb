## Try to use chunk for split the jsonl data first.
- as our file size is 14 GB and around 9 million datasize.
- use chunk size 500000 for better response

```python

chunk_size = 500000 # Number of rows to read at a time
records = [] # List to store the records
chunk_id = 0 # Initialize chunk ID
```

## Chunk the main files. called `datav17.jsonl`

```python
with open('datav17.jsonl', 'r', encoding='utf-8') as file: # Open the JSONL file for reading
    for i, line in enumerate(file): # Iterate over each line in the file
        try:
            records.append(json.loads(line))  # Attempt to parse the line as JSON and append it to the records list
        except json.JSONDecodeError:  # Handle cases where the line is not valid JSON
            continue  # Skip bad lines and proceed to the next iteration

        if (i + 1) % chunk_size == 0: # Check if the number of records has reached the chunk size
            df_chunk = pd.DataFrame(records) # Convert the records list to a DataFrame
            df_chunk.to_parquet(f"output_chunk_{chunk_id}.parquet") # Save the DataFrame to a Parquet file
            print(f"Saved chunk {chunk_id}")    # Print a message indicating the chunk has been saved
            chunk_id += 1
            records = []

    if records:
        df_chunk = pd.DataFrame(records) # Convert any remaining records to a DataFrame
        df_chunk.to_parquet(f"output_chunk_{chunk_id}.parquet") # Save any remaining records to a final chunk
        print(f"Saved chunk {chunk_id}") # Print a message indicating the final chunk has been saved
```

```python
# Function to safely parse the 'authors' column using json.loads
def safe_parse_json(x):
    try:
        # Try parsing the string as JSON (it will convert to a list/dictionary if valid)
        return json.loads(x) if isinstance(x, str) else x
    except json.JSONDecodeError:
        # If there's an error, return None or handle the error accordingly
        return None

# Apply the function to the 'authors' column
df['authors_parsed'] = df['authors'].apply(safe_parse_json)

# Check the parsed column
print(df['authors_parsed'].head())
```

# ✅ 2. Count of Papers Where Author is First Author (+ Org)
- Return: paper ID, title, author ID, name, org (only where author is first in list)

```python
first_author_rows = []

for idx, row in df_ready_clean_year.iterrows():
    if row['authors_cleaned']:
        first_author = row['authors_cleaned'][0]
        first_author_rows.append({
            'author_id': first_author.get('id'),
            'author_name': first_author.get('name'),
            'author_org': first_author.get('org'),
            'paper_id': row.get('id'),
            'title': row.get('title'),
        })

first_author_df = pd.DataFrame(first_author_rows)

# Count of first-authored papers per author
first_author_counts = first_author_df['author_id'].value_counts().reset_index()
first_author_counts.columns = ['author_id', 'first_author_paper_count']

# Merge back for detailed info
first_author_details = pd.merge(
    first_author_df, first_author_counts, on='author_id', how='left'
).drop_duplicates()

print("📌 First-author papers summary:")
print(first_author_details.head())

```

```python

# Step 2: Initialize tracking dictionary
author_type_counts = defaultdict(lambda: {
    'name': '',
    'total_conferences': 0,
    'total_journals': 0
})
# Step 3: Process each unique paper
for row in unique_papers_id.itertuples():
    paper_id = getattr(row, 'id', None)
    authors = getattr(row, 'authors_cleaned', [])
    doc_type = str(getattr(row, 'doc_type', '')).lower()  # safe conversion

    if not authors or not paper_id or doc_type not in ['conference', 'journal']:
        continue

    # First author only
    first_author = authors[0]
    aid = first_author.get('id')
    name = first_author.get('name', '')

    if not aid:
        continue

    # Count doc_type per author
    if doc_type == 'conference':
        author_type_counts[aid]['total_conferences'] += 1
    elif doc_type == 'journal':
        author_type_counts[aid]['total_journals'] += 1

    author_type_counts[aid]['name'] = name

# Step 4: Build final DataFrame
first_author_df = pd.DataFrame([
    {
        'first_author_id': aid,
        'first_author_name': data['name'],
        'total_conferences': data['total_conferences'],
        'total_journals': data['total_journals']
    }
    for aid, data in author_type_counts.items()
])

# Display preview
print("✅ Final First Author Publication Summary:")
print(first_author_df.head())

# Optional: save to file
# first_author_df.to_csv("first_author_publication_summary.csv", index=False)
```
